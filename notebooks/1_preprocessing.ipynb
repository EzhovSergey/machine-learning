# Classification

```
import pandas as pd
```

 ## Wholesale customers data

```
wcd_df = pd.DataFrame
wcd_df = pd.read_csv('wholesale_customers_data.csv' )
wcd_df.head()
```

```
wcd_df.info()
```

```
wcd_df.isnull().sum()
```

Кодирование категориального признака

```
wcd_df = pd.get_dummies(wcd_df, prefix=['Region'], columns=['Region']).rename({'Region_1': 'Region_Lisbon', 'Region_2': 'Region_Oporto', 'Region_3': 'Other Region'}, axis=1)
wcd_df.head()
```

Приведение бинарного признака к 1 или 0

```
wcd_df['Channel'] = wcd_df['Channel'].map({1: 0, 2 : 1})
wcd_df.head()
```

```
wcd_df.to_csv('wholesale_customers_data_preprocessed.csv')
```

# Regression

```
energy_df = pd.DataFrame
energy_df = pd.read_csv('energy_efficiency.csv')
energy_df
```

Удаление пустых столбцов

```
energy_df.drop(['Unnamed: 10', 'Unnamed: 11'], axis=1, inplace=True)
```

```
energy_df.head()
```

Переименование столбцов

```
columns_name = {'X1':'Compactness','X2':'S_Surface','X3':'S_Wall','X4':'S_Roof',
                'X5':'Overall_Height','X6':'Orientation','X7':'S_Glazing','X8':'Glazing Distribution',
                'Y1':'Heating Load','Y2':'Cooling Load'}

energy_df = energy_df.rename(columns_name, axis=1)
energy_df.head()
```

```
energy_df.info()
```

```
energy_df.isnull().sum()
```

Кодирование категориальных признаков

```
energy_df = pd.get_dummies(energy_df, prefix=['Orientation'], columns=['Orientation'])
energy_df = pd.get_dummies(energy_df, prefix=['Glass'], columns=['Glazing Distribution'])
energy_df.head()
```

Приведение бинарного признака к 1 или 0

```
energy_df['Overall_Height'] = energy_df['Overall_Height'].map({7.0: 1, 3.5: 0})
energy_df.head()
```

```
energy_df.to_csv('energy_efficiency_preprocessed.csv')
```
